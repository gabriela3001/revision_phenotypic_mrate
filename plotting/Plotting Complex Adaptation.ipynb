```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('NK_breaking_association_matrices_0_1.txt', 'rb') as f:
    NK_breaking_association = pickle.load(f)
with open('NK_complex_adaptation_matrices.txt', 'rb') as f:
    NK_complex_adaptation = pickle.load(f)
with open('NK_eliminating_hitchhiking_matrices.txt', 'rb') as f:
    NK_eliminating_hitchhiking = pickle.load(f)
with open('NK_limiting_pM_matrices.txt', 'rb') as f:
    NK_limiting_pM = pickle.load(f)
```

### NK $k=1$

```python
noise_labels = ['$10^{-6}$', '1.4 $\cdot$ $10^{-5}$', '2.0 $\cdot$ $10^{-4}$', '2.8 $\cdot$ $10^{-3}$', '4.0 $\cdot$ $10^{-2}$', '0.5']
simulation_names = ['Complex adaptation', 
                    'Eliminating hitchhiking',
                    'Breaking association',
                    'Limiting $p_M$']
```

```python
fig, ax = plt.subplots(4,3,figsize = (22,24))

ngens = [99,199,499]

for i in range(3):
    ax[0,i].imshow(NK_complex_adaptation[ngens[i]][0], vmin = 0, vmax = 1000)
    ax[1,i].imshow(NK_eliminating_hitchhiking[ngens[i]][0], vmin = 0, vmax = 1000)
    ax[2,i].imshow(NK_breaking_association[ngens[i]][0], vmin = 0, vmax = 1000)
    ax[3,i].imshow(NK_limiting_pM[ngens[i]][0], vmin = 0, vmax = 1000)

    
for i in range(4):
    ax[i,0].set_ylabel('Switching rate from \nnon-mutator to mutator $\gamma_1$', fontsize = 22)
    ax[i,0].text(5,75, simulation_names[i], fontsize = 24, c = 'w')
    for j in range(3):
        ax[i,j].text(5,85, 'After ' + str(ngens[j]+1) + ' generations', fontsize = 24, c = 'w')
        ax[3,j].set_xlabel('Switching rate from \nmutator to non-mutator $\gamma_2$', fontsize = 22)
        ax[i,j].set_xticks(list(np.arange(0,100)[::40]) + [99])
        ax[i,j].set_xticklabels(np.array(noise_labels)[[0,2,4,-1]], fontsize = 16)
        ax[i,j].set_yticks(list(np.arange(0,100)[::20]) + [99])
        ax[i,j].set_yticklabels(noise_labels[::-1], fontsize = 16)
        
plt.tight_layout()
#plt.savefig('complex_adaptation_NK1.jpg', dpi = 600, bbox_inches = 'tight')
```
